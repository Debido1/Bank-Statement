## DATA CLEANING

```python
# Ensure 'Debit/Credit(₦)' is numeric by removing symbols and converting
df['Debit/Credit(₦)'] = pd.to_numeric(df['Debit/Credit(₦)'].replace({',': '', r'\+': ''}, regex=True))

# Convert 'Balance(₦)' to numeric by removing any commas and ensuring proper numeric format
df['Balance(₦)'] = pd.to_numeric(df['Balance(₦)'].replace({',': ''}, regex=True))

df
```

```python
# Create 'Debit(₦)' and 'Credit(₦)' columns based on the sign
df['Debit'] = df['Debit/Credit(₦)'].apply(lambda x: abs(x) if x < 0 else 0)
df['Credit'] = df['Debit/Credit(₦)'].apply(lambda x: x if x > 0 else 0)

df
```

```python
# RENAMING CERTAIN COLUMNS
df.rename(columns={'Trans. Date':'Transaction Date','Balance(₦)':'Balance'}, inplace=True)
df
```

```python
#DROPING NECESSARY COLUMNS
df.drop(columns = ['Value Date','Debit/Credit(₦)'], inplace=True)
df
```

```python
#CONVERTING TRANSACTION DATE TO DATE TIME FORMAT
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'] , errors='coerce')
df
```

```python
#EXTRACTING THE MONTH FROM THE TRANSACTION DATE
df['Month']= df['Transaction Date'].dt.month_name()
df
```

## DATA ANALYSIS

```python
channel_summary = df['Channel'].value_counts()
print(channel_summary)
```

```python
total_channel_summary = df['Channel'].count()
print(total_channel_summary)
```

```python
#FINDING TOTAL INCOME AND EXPENSES
total_income = df['Credit'].sum()

total_expenses = df['Debit'].sum()

print('The total income is {}'.format(total_income))
print('The total expenses is {}'.format(total_expenses))
```

```python
# Time Series Analysis: Use Transaction Date to analyze trends over time, such as monthly expenses or income trends.
monthly_expenses = df.resample('M', on='Transaction Date')['Debit'].sum()

monthly_income = df.resample('M', on='Transaction Date')['Credit'].sum()

print(monthly_expenses)
print(monthly_income)
```

```python
df
```

```python
df['total_balance'] = df['Credit'] - df['Debit']
df
```

```python
channel_ssummary = df.groupby('Channel')['Debit'].sum()
print(channel_ssummary)
```

```python
balance = df['total_balance'].sum()
print("Total Balance:", balance)
```

```python
#Outlier Detection: Identifing unusually high transactions in Debit or credit as potential outliers
high_transactions = df[df['Debit'] > df['Debit'].quantile(0.95)]
high_transactions
```

## DATA VISUALIZATION

```python
df.plot(x='Month', y='Debit', kind='line')

# Plot Monthly Expenses and Income
plt.figure(figsize=(10, 5))
monthly_expenses.plot(label='Monthly Expenses', color='red')
monthly_income.plot(label='Monthly Income', color='green')
plt.legend()
plt.title('Monthly Expenses and Income')
plt.show()
```

```python
# Bar Chart for Frequency
plt.figure(figsize=(6, 5))
channel_summary.plot(kind='bar', color='lightcoral')
plt.title('Transaction Frequency by Channel')
plt.xlabel('Channel')
plt.ylabel('Number of Transactions')
plt.show()

# Bar Chart for Spending by Channel
plt.figure(figsize=(6, 5))
channel_ssummary.plot(kind='bar', color='steelblue')
plt.title('Total Spending by Channel')
plt.xlabel('Channel')
plt.ylabel('Amount Spent')
plt.show()
```

```python
# Plot Daily Balance
plt.figure(figsize=(11, 6))
plt.plot(df['Transaction Date'], df['Balance'], marker='o', color='purple')
plt.title('Daily Balance Trend')
plt.xlabel('Date')
plt.ylabel('Balance')
plt.grid(True)
plt.show()
```
